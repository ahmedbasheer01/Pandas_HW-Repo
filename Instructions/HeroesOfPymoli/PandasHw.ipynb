## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Remove duplicate names
duplicate = purchase_data.drop_duplicates(subset='SN', keep="first")

duplicate = duplicate.sort_values(by=['Age'], ascending=True)
duplicate.head(20)

# Create the bins
bins = [0, 10, 15, 20, 25, 30, 35, 40, 150]

# Create the names for the bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

duplicate["Age Group"] = pd.cut(duplicate['Age'], bins, labels=group_names, right=False)
duplicate.head(2)
```

```python
demographics = pd.DataFrame(duplicate["Age Group"].value_counts())
demographics.sort_index(inplace=True)

demographics= demographics.rename(columns={'Age Group':'Total Count'})
tot = demographics['Total Count'].sum()
demographics['Percentage'] = (100*demographics['Total Count']/tot)

demographics['Percentage'] = demographics['Percentage'].map('{:,.2f}%'.format)
demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Age Group"] = pd.cut(purchase_data['Age'], bins, labels=group_names, right=False)
purchase_data.head(6)
```

```python
#Avg
demographics2 = pd.DataFrame(duplicate["Age Group"].value_counts())
demographics2 = demographics2.reset_index()
demographics2 = demographics2.rename(columns={'index':'Age Group', 'Age Group':'Total Indiv Count'})
demographics2 = demographics2.set_index('Age Group')
demographics2.head(20)
```

```python
Age_purchase = pd.DataFrame(purchase_data["Age Group"].value_counts())
Age_purchase = Age_purchase.reset_index()
Age_purchase = Age_purchase.rename(columns={'index': 'Age Group', 'Age Group':'Total Count'})
Age_purchase = Age_purchase.set_index('Age Group')

#Average Purchase Price
purchase_data_group = purchase_data.groupby('Age Group').mean()
purchase_data_group1 = pd.DataFrame(purchase_data_group['Price'])

#Total Purchases
purchase_data_tpv = purchase_data.groupby('Age Group').sum()
purchase_data_group2 = pd.DataFrame(purchase_data_tpv['Price'])


#Merge Data
Age_group_table = pd.merge(Age_purchase, purchase_data_group1, on="Age Group", how="outer")
Age_group_table = pd.merge(Age_group_table, purchase_data_group2, on="Age Group", how="outer")
Age_group_table = pd.merge(Age_group_table, demographics2, on="Age Group", how="outer")
Age_group_table ['Avg Total Purchase per Person'] = (Age_group_table['Price_y']/Age_group_table['Total Indiv Count'])
Age_group_table = Age_group_table.drop('Total Indiv Count', 1)

#Format Data
Age_group_table = Age_group_table.rename(columns={'Total Count': 'Purchase Count',\
                                                  'Price_x':'Average Purchase Price',\
                                                  'Price_y':'Total Purchase Value'})

Age_group_table['Average Purchase Price'] = Age_group_table['Average Purchase Price'].map('${:,.2f}'.format)
Age_group_table['Total Purchase Value'] = Age_group_table['Total Purchase Value'].map('${:,.2f}'.format)
Age_group_table['Avg Total Purchase per Person'] = Age_group_table['Avg Total Purchase per Person'].map('${:,.2f}'.format)

Age_group_table = Age_group_table.sort_values(by=['Age Group'], ascending=True)
Age_group_table
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_data_top = purchase_data.groupby(['SN']).sum()#sum of purchases
purchase_data_top=purchase_data_top.sort_values('Price', ascending=False )#sorted by sum of spent
purchase_data_top = purchase_data_top[['Price']]


purchase_data_top1 = purchase_data.groupby(['SN']).count()#count of purchases
purchase_data_top1=purchase_data_top1.sort_values('Purchase ID', ascending=False )#sorted by number of purchases
purchase_data_top1 = purchase_data_top1[['Purchase ID']]

topspenderstable = pd.merge(purchase_data_top1, purchase_data_top, on='SN', how='outer')
topspenderstable['Avg Price'] = topspenderstable['Price'] / topspenderstable['Purchase ID']

#FORMATTING
topspenderstable['Price'] = topspenderstable['Price'].map('${:,.2f}'.format)
topspenderstable['Avg Price'] = topspenderstable['Avg Price'].map('${:,.2f}'.format)
topspenderstable = topspenderstable[['Purchase ID','Avg Price','Price']]
topspenderstable = topspenderstable.rename(columns={'Purchase ID':'Purchase Count',\
                                                    'Avg Price':'Average Purchase Price',\
                                                    'Price':'Total Purchase Price'})
                                                    
topspenderstable.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_data.head(3)
```

```python
#RETRIEVE ITEMS DATA
purchase_data_items = purchase_data[['Item ID', 'Item Name', 'Price']]

#COUNT PURCHASES
gp_items = purchase_data_items.groupby(['Item ID', 'Item Name'],as_index=False).count()
gp_items= gp_items.rename(columns={'Price':'Purchase Count'})

#SUM PRICE
gp_items1 = purchase_data_items.groupby(['Item ID', 'Item Name'],as_index=False).sum()

#MERGE TABLES
sum_table_items = pd.merge(gp_items,gp_items1,on='Item ID', how = 'outer')
sum_table_items= sum_table_items.rename(columns={'Price':'Total Purchase Price'})

#CREATE PRICE COLUMN
sum_table_items['Price'] = round((sum_table_items['Total Purchase Price'] / sum_table_items['Purchase Count']),2)

#REMOVE UN-NEEDED COLUMNS
sum_table_items = sum_table_items[['Item ID', 'Item Name_x', 'Purchase Count','Price', 'Total Purchase Price']]
sum_table_items= sum_table_items.rename(columns={'Item Name_x':'Item Name'})

#SORT
sum_table_items=sum_table_items.sort_values('Purchase Count', ascending=False )#sorted by purchase count

#SET INDEX
sum_table_items = sum_table_items.set_index('Item ID')

#FORMATTING
sum_table_items['Price'] = sum_table_items['Price'].map('${:,.2f}'.format)
sum_table_items['Total Purchase Price'] = sum_table_items['Total Purchase Price'].map('${:,.2f}'.format)
sum_table_items.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
sum_table_profits=sum_table_items

sum_table_profits['Total Purchase Price'] = sum_table_profits['Total Purchase Price'].str.replace('$', '')
#df['purchase'].astype(str).astype(int)
sum_table_profits['Total Purchase Price'] = sum_table_profits['Total Purchase Price'].astype(float)

#SORT
sum_table_profits=sum_table_profits.sort_values('Total Purchase Price', ascending=False )#sorted by sum of spent

#FORMATTING
sum_table_profits['Total Purchase Price'] = sum_table_profits['Total Purchase Price'].map('${:,.2f}'.format)
sum_table_profits.head()
```
